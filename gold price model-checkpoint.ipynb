```python
## spliting the data and label
```

```python
x=gold_data.drop(['Date','GLD'],axis=1)
y=gold_data['GLD']
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=2 )
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#training of model:random forest regressor
```

```python
regressor= RandomForestRegressor(n_estimators=100)
```

```python
#training the model
```

```python
regressor.fit(x_train,y_train)
```

```python
#prediction of data of model
```

```python
data_type=regressor.predict(x_test)
```

```python
print(data_type)
```

```python
# R squared error
error_score = metrics.r2_score(y_test, data_type)
print("R squared error : ", error_score)
```

```python
#compare actual and predict value
y_test = list(y_test)
```

```python
plt.plot(y_test, color='blue', label = 'Actual Value')
plt.plot(data_type, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
